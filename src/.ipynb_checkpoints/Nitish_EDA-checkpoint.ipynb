```python
def preprocess(ReviewText):
    ReviewText = ReviewText.replace("(<br/>)", "")
    ReviewText = ReviewText.replace('(<a).*(>).*(</a>)', '')
    ReviewText = ReviewText.replace('(&amp)', '')
    ReviewText = ReviewText.replace('(&gt)', '')
    ReviewText = ReviewText.replace('(&lt)', '')
    ReviewText = ReviewText.replace('(\xa0)', ' ')  
    return ReviewText
```

```python
def process_text(text,flag=False):
    stop_words = stopwords.words('english')
    text = text.lower()
    text = text.replace('\n','')
    text = re.sub(r'\(.*\)','',text)
    text = re.sub(r'[^a-zA-Z0-9. ]','',text)
    text = re.sub(r'\.',' . ',text)
    text = text.replace('.','')
    text = text.split()
    for i in range(len(text)):
        word = text[i]
        if word in contraction_mapping:
            text[i] = contraction_mapping[word]
    newtext = []
    for word in text:
        if word not in stop_words and len(word)>0:
            newtext.append(word)
    text = newtext
    if flag:
        text = text[::-1]
    text = " ".join(text)
    text = text.replace("'s",'') 
    return text

def get_train_test_data(source,target):
    corpus_source = open(source, 'r').readlines()
    corpus_target = open(target, 'r').readlines()
    return corpus_source,corpus_target

def prepareInput(sourcefilePath,destFilePath,fileName):
    source = sourcefilePath
    target = destFilePath
#     save_trans = fileName

    train_words_list = []
    test_words_list = []
        
    corpus_source,corpus_target = get_train_test_data(source,target)
    
#     corpus_source = open(source, 'r').readlines()
#     corpus_target = open(target, 'r').readlines()
  
#     writer = open(save_trans, 'w')
    num_lines = 0
    for k, v in zip(corpus_source, corpus_target):
#         k = process_text(k,True)
#         v = process_text(v,False)
#         if num_lines > 10:
#             break
#         print("Source ",k)
#         print("Target ",v)
        splitted_train = k.split(' ')
        splitted_test = v.split(' ')
        for item in splitted_train:
#             item = re.sub('\W+','', item)
            item = preprocess(item)
            if item=='':
                continue
            train_words_list.append(item)
        for item in splitted_test:
#             item = re.sub('\W+','', item)
            item = preprocess(item)
            if item=='':
                continue
            test_words_list.append(item)
        num_lines = num_lines + 1
#         writer.write(k + '\t' + v + '\n')
#         num_lines += 1
#         if num_lines > MAX_SIZE:
#             break
#     writer.flush()
#     writer.close()
    return train_words_list,test_words_list
```

```python
def getOOVwords(train_list,test_list):
    oov_list = []
    for item_test in test_list:
        if item_test not in train_list:
            if item_test not in oov_list:
                oov_list.append(item_test)
    return oov_list
```

```python
def length_of_para_fre(corpus_source, corpus_target):
    length_of_para_train = []
    len_of_para_summary = []
    for k, v in zip(corpus_source, corpus_target):
        splitted_train = k.split(' ')
        splitted_test = v.split(' ')
        if len(splitted_train) < 200:
            length_of_para_train.append(len(splitted_train))
            len_of_para_summary.append(len(splitted_test))
#         length_of_para.append(len(splitted_train))
    return length_of_para_train,len_of_para_summary



length_of_para_train,len_of_para_summary = length_of_para_fre(corpus_source,corpus_target)

length_of_keys = Counter(length_of_para_train).keys()
frequency_of_keys = Counter(length_of_para_train).values()

print("length_of_keys",length_of_keys,"")
print("frequency_of_keys",frequency_of_keys,"")

length_of_keys_summary = Counter(len_of_para_summary).keys()
frequency_of_keys_summary = Counter(len_of_para_summary).values()
```

```python
def Bar_Plot(length_of_keys,frequency_of_keys,name):
    import matplotlib.pyplot as plt
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    langs = length_of_keys
    students = frequency_of_keys
    ax.bar(langs,students)
    plt.title(name + " Length v/s Frequency")
    plt.show()
    plt.savefig("Bar_Plot_Length_Frequency_"+name+".png")
```

```python
def Scatter_PLot(length_of_keys,frequency_of_keys,name):
    plt.title(name + "Word Frequencies ")
    plt.ylabel("Total Number of Occurrences")
    plt.xlabel("Rank")
    plt.scatter(
    length_of_keys,
    frequency_of_keys,
    color="blue"
    )
    plt.show()
    plt.savefig("Scatter_Plot_Length_Frequency_"+name+".png")
```

```python
Scatter_PLot(length_of_keys,frequency_of_keys,'Paragraph_')
Scatter_PLot(length_of_keys_summary,frequency_of_keys_summary,'Summary_')

Bar_Plot(length_of_keys,frequency_of_keys,'Paragraph_')
Bar_Plot(length_of_keys_summary,frequency_of_keys_summary,'Summary_')
```
